```python
# Shape and info
df.shape
df.info()

# Missing values
df.isnull().sum()

# Summary stats
df.describe(include='all')
```

```python
# Convert dates if necessary
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Add Revenue and Profit
df['Revenue'] = df['Selling Unit Price (Pre Discount)'] * df['Quantity']
df['Cost'] = df['Purchase Unit Cost'] * df['Quantity']
df['Profit'] = df['Revenue'] - df['Cost']

# Profit margin
df['Profit Margin %'] = (df['Profit'] / df['Revenue']) * 100
```

```python
# Histograms for numeric features
num_cols = ['Quantity', 'Discount%', 'Purchase Unit Cost', 'Selling Unit Price (Pre Discount)', 'Profit', 'Revenue']
df[num_cols].hist(bins=20, figsize=(12, 10), color='skyblue')
plt.suptitle("Histograms of Numeric Features")
plt.show()
# Histograms for numeric features
num_cols = ['Quantity', 'Discount%', 'Purchase Unit Cost', 'Selling Unit Price (Pre Discount)', 'Profit', 'Revenue']
df[num_cols].hist(bins=20, figsize=(12, 10), color='skyblue')
plt.suptitle("Histograms of Numeric Features")
plt.show()
```

```python
cat_cols = ['Category', 'Sub-Category', 'Region', 'Segment']

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Profit by Category
plt.figure(figsize=(8,5))
sns.boxplot(x='Category', y='Profit', data=df)
plt.title('Profit by Category')
plt.show()

# Revenue by Region
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Revenue', data=df, estimator=sum, ci=None)
plt.title('Total Revenue by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
# Revenue over time
df_by_date = df.groupby('Order Date')['Revenue'].sum().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(data=df_by_date, x='Order Date', y='Revenue')
plt.title("Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Heatmap of numerical features
corr = df[['Quantity', 'Purchase Unit Cost', 'Selling Unit Price (Pre Discount)', 'Discount%', 'Revenue', 'Profit']].corr()

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
